```python
from pdf2image import convert_from_path
import easyocr
import numpy as np
```

```python
import PIL
from PIL import ImageDraw
```

```python
import spacy
```

```python

#enter download link
Link = input("Enter the pdf Link and Press Enter - ")

reader = easyocr.Reader(['en'])

import requests

#download link and transfeering to sample.pdf

url = 'http://www.africau.edu/images/default/sample.pdf'
r = requests.get(url, allow_redirects=True)

open('sample.pdf', 'wb').write(r.content)
```

```python

images = convert_from_path('sample.pdf',poppler_path='poppler-21.11.0/Library/bin')
```

```python
from IPython.display import display,Image
```

```python
#displaying file
display (images[0])
```

```python
# drawing rectangle around text

bounds = reader.readtext(np.array(images[0]))
bounds
def draw_boxes(image , bounds ,color = 'yellow',width=2):
    draw = ImageDraw.Draw(image)
    for bound in bounds:
        p0,p1,p2,p3 = bound[0]
        draw.line([*p0,*p1,*p2,*p3,*p0],fill = color ,width = width)
        
    return image
draw_boxes(images[0] ,bounds)
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
#extraction of text
text = ''
for i in range(len(bounds)):
    text = text + bounds[i][1] + '\n';
doc = nlp(text)
for i in text.split():
    print(i)
```
